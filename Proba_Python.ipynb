# AGUJA DE BUFFON
Una manera que conocemos para calcular el valor de  es trazando un círculo y dividiendo lo que mide su circunferencia entre lo que mide su diámetro. 
Sin embargo, desde hace cientos de años, los matemáticos han desarrollado otras maneras para llegar al número. 
Una de ellas es el experimento propuesto por el Conde de Buffon » en 1777.
Este experimento consiste en dejar caer una aguja sobre una hoja rayada y anotar las veces que la aguja cruza alguna de las rayas. Después de lanzar la aguja muchísimas veces comprobó que su experimento estaba íntimamente relacionado con el número . Para obtener un número muy parecido a , hay que dejar caer la aguja muchísimas veces sobre la hoja, multiplicar esta cantidad por dos y dividir el resultado entre el número de veces que la aguja cruzó alguna de las rayas.
Con el llamado "problema de la aguja de Buffon", (siglo XVIII ), nace la teoría de las probabilidades geométricas desarrollada poco después por Laplace » en su gran tratado "Teoría analítica de las probabilidades" (1812).

 $$ \pi = \frac{2* número-de-lanzamientos}{número-de-veces-que-tocó-una-línea}$$

# BARAJA

Un mazo consta de 52 cartas, 13 corazones, 13 tréboles, 13 ases y 13 rombos. 
¿cuál es la probabilidad de que al barajear un mazo de cartas, salgan cuatro ases seguidos?

$$\frac{n! \cdot (m-n+1)!}{m!}$$

- $n = 4$ son los 4 ases. Hay un as de corazones, uno de tréboles y uno de rombos. Éstos pueden aparecer el cualquier posición, en realidad, en $4! = 24$ posiciones. 

- $n! \cdot (m-n+1)!$ representa que, éstos 4 ases pueden aparecer a lo largo de toda la baraja, recorriendola, pueden aparecer en $(m-n+1)!$ formas

- Al dividirlo entre $m! = 52!$ estamos recorriendo todas las cartas del mazo

## Factorial

Definimos la función factorial de la siguiente manera:

```python
def factorial(n):
    #caso base
    if n == 0:
        return 1
    else:
        return factorial(n-1)*n 
```

```python
import random as rd
import numpy as np
#import numba as nb
```

```python
n=1000000 #barajeando el mazo un millón de veces
deck = [0,0,0,0] + list(range(1,49)) #los ceros son nuestros ases
cuenta = 0

for i in range(n):
    rd.shuffle(deck)

    for i in range(48):
        if [deck[i], deck[i+1], deck[i+2], deck[i+3]] == [0,0,0,0]: #si hay cuatro ases seguidos
            cuenta += 1
            
print(cuenta/n)
```

```python
a = cuenta/n
```

```python
print("La probabilidad experimental es de: ", a)
```

```python
n = 4
m = 52
```

```python
proba = ( factorial(n) * factorial(m-n+1) ) / factorial(m)
```

```python
print("La probabilidad teórica es de: ", proba) 
```

```python
error = (a - proba) / a
```

```python
print("el error porcentual entre la probabilidad teórica y la experimental es de: ", error*100, "%")
```
